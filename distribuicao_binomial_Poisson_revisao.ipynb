# Heredidade mendeliana

Cruzamento Aa x Aa - probabilidade de descendente recessivo 10 descendentes, prob. de aa = 0.25


```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parâmetros
n = 10  # número de tentativas
p = 0.25  # probabilidade de sucesso

# Criar distribuição binomial
dist = binom(n=n, p=p)

# Calcular PMF para k específico
k = 2
pmf_k = dist.pmf(k)
print(f'P(X = {k}) = {pmf_k:.4f}')

# Calcular CDF para k específico
cdf_k = dist.cdf(k)
print(f'P(X ≤ {k}) = {cdf_k:.4f}')

# Média e variância
media = dist.mean()  # = 2.50
variancia = dist.var()  # = 1.88
print(f'Média: {media:.2f}')
print(f'Variância: {variancia:.2f}')

# Gerar valores para plotar
x = np.arange(0, n+1)
pmf_values = dist.pmf(x)
cdf_values = dist.cdf(x)

# Plotar PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(x, pmf_values, alpha=0.7)
plt.title('PMF - Distribuição Binomial')
plt.xlabel('k (número de sucessos)')
plt.ylabel('P(X = k)')

# Plotar CDF
plt.subplot(1, 2, 2)
plt.step(x, cdf_values, where='post')
plt.title('CDF - Distribuição Binomial')
plt.xlabel('k')
plt.ylabel('P(X ≤ k)')
plt.show()
```


# Número de bactérias por ml

Contagem de bactérias E.coli em cultura Média de 3.5 bactérias por ml


```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parâmetros
lam = 3.5  # taxa média (lambda)

# Criar distribuição de Poisson
dist = poisson(mu=lam)

# Calcular PMF para k específico
k = 3
pmf_k = dist.pmf(k)
print(f'P(X = {k}) = {pmf_k:.4f}')

# Calcular CDF para k específico
cdf_k = dist.cdf(k)
print(f'P(X ≤ {k}) = {cdf_k:.4f}')

# Média e variância
media = dist.mean()  # = 3.50
variancia = dist.var()  # = 3.50
print(f'Média: {media:.2f}')
print(f'Variância: {variancia:.2f}')

# Gerar valores para plotar
x = np.arange(0, int(lam*3)+1)
pmf_values = dist.pmf(x)
cdf_values = dist.cdf(x)

# Plotar PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(x, pmf_values, alpha=0.7)
plt.title('PMF - Distribuição de Poisson')
plt.xlabel('k (número de eventos)')
plt.ylabel('P(X = k)')

# Plotar CDF
plt.subplot(1, 2, 2)
plt.step(x, cdf_values, where='post')
plt.title('CDF - Distribuição de Poisson')
plt.xlabel('k')
plt.ylabel('P(X ≤ k)')
plt.show()
```
