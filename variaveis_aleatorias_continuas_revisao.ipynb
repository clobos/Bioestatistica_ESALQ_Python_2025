# Altura de plantas (cm)

Altura de uma espécie de planta herbácea

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parâmetros da distribuição normal
media = 25
desvio = 5

# Criar distribuição normal
dist = stats.norm(loc=media, scale=desvio)

# Definir intervalo para plotagem
x = np.linspace(10, 40, 1000)

# Calcular PDF e CDF
pdf_values = dist.pdf(x)
cdf_values = dist.cdf(x)

# Exemplos de cálculo
a, b = 20, 30
prob_intervalo = dist.cdf(b) - dist.cdf(a)

print(f"Média: {media}")
print(f"Desvio padrão: {desvio}")
print(f"P({a} ≤ X ≤ {b}) = {prob_intervalo:.4f}")

# Plotar PDF e CDF
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, linewidth=2, color='#B4413C')
plt.fill_between(x[(x >= a) & (x <= b)],
                 pdf_values[(x >= a) & (x <= b)],
                 alpha=0.3, color='#B4413C')
plt.xlabel('Altura de plantas ')
plt.ylabel('Densidade f(x)')
plt.title('Função Densidade de Probabilidade (PDF)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, linewidth=2, color='#5D878F')
plt.axhline(y=dist.cdf(a), color='red', linestyle='--', alpha=0.7)
plt.axhline(y=dist.cdf(b), color='red', linestyle='--', alpha=0.7)
plt.xlabel('Altura de plantas ')
plt.ylabel('F(x) = P(X ≤ x)')
plt.title('Função de Distribuição Acumulada (CDF)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Calcular probabilidades específicas
print(f"\nExemplos de cálculo:")
print(f"P(X ≤ {media}) = {dist.cdf(media):.4f}")
print(f"P(X > {media + desvio}) = {1 - dist.cdf(media + desvio):.4f}")
print(f"P({media - desvio} ≤ X ≤ {media + desvio}) = {dist.cdf(media + desvio) - dist.cdf(media - desvio):.4f}")
```
