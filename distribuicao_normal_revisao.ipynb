# Exemplo: Peso corporal (kg)

Peso de uma espécie de mamífero


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parâmetros da distribuição
mu = 45  # média
sigma = 8  # desvio padrão

# Criar distribuição normal
dist = stats.norm(mu, sigma)

# Gerar valores para plotagem
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calcular PDF e CDF
pdf_values = dist.pdf(x)
cdf_values = dist.cdf(x)

# Plotar gráficos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# PDF
ax1.plot(x, pdf_values, 'b-', linewidth=2, label='PDF')
ax1.fill_between(x, pdf_values, alpha=0.3)
ax1.axvline(mu, color='r', linestyle='--', label='Média')
ax1.set_xlabel('Peso corporal (kg)')
ax1.set_ylabel('Densidade')
ax1.set_title('Função Densidade de Probabilidade')
ax1.legend()
ax1.grid(True, alpha=0.3)

# CDF
ax2.plot(x, cdf_values, 'g-', linewidth=2, label='CDF')
ax2.axhline(0.5, color='r', linestyle=':', label='P = 0.5')
ax2.set_xlabel('Peso corporal (kg)')
ax2.set_ylabel('Probabilidade Acumulada')
ax2.set_title('Função de Distribuição Acumulada')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Cálculos úteis
print(f"Média: {mu}")
print(f"Desvio padrão: {sigma}")
print(f"P(X ≤ {mu}): {dist.cdf(mu):.4f}")
print(f"P({mu-sigma} ≤ X ≤ {mu+sigma}): {dist.cdf(mu+sigma) - dist.cdf(mu-sigma):.4f}")

# Percentis
print(f"Percentil 25: {dist.ppf(0.25):.2f}")
print(f"Percentil 50: {dist.ppf(0.50):.2f}")
print(f"Percentil 75: {dist.ppf(0.75):.2f}")
```
