# RESPOSTAS - VARIÁVEIS ALEATÓRIAS DISCRETAS

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

# ==========================================
# FUNÇÕES AUXILIARES
# ==========================================

def verificar_funcao_probabilidade(valores, probabilidades):
    """
    Verifica se uma função é válida como função de massa de probabilidade
    """
    # Verificar se todas as probabilidades são não-negativas
    todas_positivas = all(p >= 0 for p in probabilidades)

    # Verificar se a soma é 1 (com tolerância para erros de arredondamento)
    soma = sum(probabilidades)
    soma_eh_um = abs(soma - 1.0) < 1e-10

    print(f"Todas as probabilidades ≥ 0: {todas_positivas}")
    print(f"Soma das probabilidades: {soma:.6f}")
    print(f"Soma = 1: {soma_eh_um}")
    print(f"É função de probabilidade válida: {todas_positivas and soma_eh_um}")

    return todas_positivas and soma_eh_um

def criar_tabela_distribuicao(valores, probabilidades):
    """
    Cria tabela da distribuição da variável aleatória
    """
    # Calcular FDA
    fda = np.cumsum(probabilidades)

    # Criar DataFrame
    df = pd.DataFrame({
        'X = x': valores,
        'P(X = x)': probabilidades,
        'F(x) = P(X ≤ x)': fda
    })

    print("\nTabela da Distribuição:")
    print(df.to_string(index=False, float_format='%.4f'))
    return df

def calcular_momentos(valores, probabilidades):
    """
    Calcula esperança e variância
    """
    # Esperança E[X]
    esperanca = sum(x * p for x, p in zip(valores, probabilidades))

    # E[X²]
    esperanca_x2 = sum(x**2 * p for x, p in zip(valores, probabilidades))

    # Variância Var[X] = E[X²] - [E[X]]²
    variancia = esperanca_x2 - esperanca**2

    # Desvio padrão
    desvio_padrao = np.sqrt(variancia)

    print(f"\nMomentos:")
    print(f"E[X] = {esperanca:.4f}")
    print(f"E[X²] = {esperanca_x2:.4f}")
    print(f"Var[X] = {variancia:.4f}")
    print(f"DP[X] = {desvio_padrao:.4f}")

    return esperanca, variancia, desvio_padrao

def fazer_graficos(valores, probabilidades, titulo):
    """
    Cria gráficos da função de probabilidade e FDA
    """
    fda = np.cumsum(probabilidades)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Gráfico da função de probabilidade
    ax1.bar(valores, probabilidades, alpha=0.7, color='skyblue', edgecolor='black')
    ax1.set_xlabel('X')
    ax1.set_ylabel('P(X = x)')
    ax1.set_title(f'Função de Probabilidade\n{titulo}')
    ax1.grid(True, alpha=0.3)

    # Gráfico da FDA
    ax2.step(valores, fda, where='post', linewidth=2, color='red')
    ax2.scatter(valores, fda, color='red', s=50, zorder=5)
    ax2.set_xlabel('X')
    ax2.set_ylabel('F(x) = P(X ≤ x)')
    ax2.set_title(f'Função de Distribuição Acumulada\n{titulo}')
    ax2.grid(True, alpha=0.3)
    ax2.set_ylim(0, 1.1)

    plt.tight_layout()
    plt.show()

# ==========================================
# EXERCÍCIO 1 - RESPOSTA
# ==========================================
print("=" * 50)
print("EXERCÍCIO 1: Ovos viáveis em ninhada")
print("=" * 50)

# Dados do problema
n = 3  # número de ovos
p = 0.8  # probabilidade de viabilidade
x_vals = [0, 1, 2, 3]

# Calcular probabilidades usando distribuição binomial
probabilidades = [stats.binom.pmf(k, n, p) for k in x_vals]

print("Função dada: P(X = k) = C(3,k) * (0.8)^k * (0.2)^(3-k)")
print("\nCálculo das probabilidades:")
for i, (x, prob) in enumerate(zip(x_vals, probabilidades)):
    print(f"P(X = {x}) = C(3,{x}) * (0.8)^{x} * (0.2)^{3-x} = {prob:.4f}")

# 1. Verificar se é função de probabilidade
print("\n1. Verificação se é função de probabilidade:")
verificar_funcao_probabilidade(x_vals, probabilidades)

# 2. Criar tabela
print("\n2. Tabela da distribuição:")
tabela = criar_tabela_distribuicao(x_vals, probabilidades)

# 3. e 4. Fazer gráficos
print("\n3. e 4. Gráficos:")
fazer_graficos(x_vals, probabilidades, "Ovos viáveis")

# 5. e 6. Calcular momentos
print("\n5. e 6. Momentos:")
esp, var, dp = calcular_momentos(x_vals, probabilidades)

print("\nInterpretação biológica:")
print(f"- Esperamos em média {esp:.2f} ovos viáveis por ninhada")
print(f"- A variabilidade é de {dp:.2f} ovos (desvio padrão)")
print(f"- Como p={p} é alto, a maioria das ninhadas terá 2-3 ovos viáveis")

# ==========================================
# EXERCÍCIO 2 - RESPOSTA
# ==========================================
print("\n" + "=" * 50)
print("EXERCÍCIO 2: Mutações em gene")
print("=" * 50)

x_vals = [0, 1, 2, 3, 4]
probabilidades = [0.4, 0.3, 0.2, 0.08, 0.02]

print("Probabilidades dadas:")
for x, p in zip(x_vals, probabilidades):
    print(f"P(X = {x}) = {p}")

# 1. Verificar se é função de probabilidade
print("\n1. Verificação se é função de probabilidade:")
verificar_funcao_probabilidade(x_vals, probabilidades)

# 2. Criar tabela
print("\n2. Tabela da distribuição:")
tabela = criar_tabela_distribuicao(x_vals, probabilidades)

# 3. e 4. Fazer gráficos
print("\n3. e 4. Gráficos:")
fazer_graficos(x_vals, probabilidades, "Mutações em gene")

# 5. e 6. Calcular momentos
print("\n5. e 6. Momentos:")
esp, var, dp = calcular_momentos(x_vals, probabilidades)

print("\nInterpretação biológica:")
print(f"- Esperamos em média {esp:.2f} mutações por gene")
print(f"- A probabilidade de não haver mutação é {probabilidades[0]*100}%")
print(f"- Genes com muitas mutações (≥3) são raros ({(probabilidades[3]+probabilidades[4])*100}%)")

# ==========================================
# EXERCÍCIO 3 - RESPOSTA
# ==========================================
print("\n" + "=" * 50)
print("EXERCÍCIO 3: Sobrevivência de filhotes")
print("=" * 50)

n = 4  # número de filhotes
p = 0.75  # probabilidade de sobrevivência
x_vals = list(range(n + 1))

# Calcular probabilidades
probabilidades = [stats.binom.pmf(k, n, p) for k in x_vals]

print(f"Distribuição Binomial: n = {n}, p = {p}")
print("\nCálculo das probabilidades:")
for x, prob in zip(x_vals, probabilidades):
    print(f"P(X = {x}) = C(4,{x}) * (0.75)^{x} * (0.25)^{4-x} = {prob:.4f}")

# Executar todas as análises
verificar_funcao_probabilidade(x_vals, probabilidades)
tabela = criar_tabela_distribuicao(x_vals, probabilidades)
fazer_graficos(x_vals, probabilidades, "Sobrevivência de filhotes")
esp, var, dp = calcular_momentos(x_vals, probabilidades)

print("\nValidação teórica (Binomial):")
print(f"E[X] teórico = n*p = {n}*{p} = {n*p}")
print(f"Var[X] teórico = n*p*(1-p) = {n}*{p}*{1-p} = {n*p*(1-p)}")

# ==========================================
# EXERCÍCIO 4 - RESPOSTA
# ==========================================
print("\n" + "=" * 50)
print("EXERCÍCIO 4: Células infectadas")
print("=" * 50)

x_vals = [0, 1, 2, 3, 4, 5]
probabilidades = [0.32, 0.40, 0.20, 0.06, 0.015, 0.005]

# Executar análises completas
verificar_funcao_probabilidade(x_vals, probabilidades)
tabela = criar_tabela_distribuicao(x_vals, probabilidades)
fazer_graficos(x_vals, probabilidades, "Células infectadas")
esp, var, dp = calcular_momentos(x_vals, probabilidades)

print("\nAnálise epidemiológica:")
prob_poucas_infeccoes = sum(probabilidades[:3])  # 0, 1, 2 células
print(f"P(≤ 2 células infectadas) = {prob_poucas_infeccoes:.3f}")
print(f"Isso indica baixa transmissibilidade do vírus")

# ==========================================
# EXERCÍCIO 5 - RESPOSTA
# ==========================================
print("\n" + "=" * 50)
print("EXERCÍCIO 5: Espécies de fungos")
print("=" * 50)

x_vals = [1, 2, 3, 4, 5]
probabilidades = [0.1, 0.25, 0.35, 0.25, 0.05]

# Executar análises
verificar_funcao_probabilidade(x_vals, probabilidades)
tabela = criar_tabela_distribuicao(x_vals, probabilidades)
fazer_graficos(x_vals, probabilidades, "Espécies de fungos")
esp, var, dp = calcular_momentos(x_vals, probabilidades)

print("\nAnálise de biodiversidade:")
print(f"Valor modal: 3 espécies (P = {max(probabilidades):.2f})")
print(f"Diversidade média: {esp:.2f} espécies por amostra")

# ==========================================
# TEMPLATE PARA DEMAIS EXERCÍCIOS
# ==========================================
print("\n" + "=" * 50)
print("TEMPLATE PARA EXERCÍCIOS 6-20")
print("=" * 50)

def resolver_exercicio_discreto(titulo, x_vals, probabilidades_ou_funcao, parametros=None):
    """
    Template para resolver qualquer exercício de variável discreta

    Args:
        titulo: Nome do exercício
        x_vals: Lista de valores possíveis
        probabilidades_ou_funcao: Lista de probabilidades ou nome da distribuição
        parametros: Parâmetros se for distribuição conhecida
    """
    print(f"\n{'=' * 50}")
    print(f"{titulo}")
    print(f"{'=' * 50}")

    # Se for lista de probabilidades
    if isinstance(probabilidades_ou_funcao, list):
        probabilidades = probabilidades_ou_funcao
    # Se for distribuição conhecida
    elif probabilidades_ou_funcao == 'binomial':
        n, p = parametros
        probabilidades = [stats.binom.pmf(k, n, p) for k in x_vals]
        print(f"Distribuição Binomial: n = {n}, p = {p}")
    elif probabilidades_ou_funcao == 'poisson':
        lambda_val = parametros
        probabilidades = [stats.poisson.pmf(k, lambda_val) for k in x_vals]
        print(f"Distribuição Poisson: λ = {lambda_val}")

    # Executar todas as análises
    verificar_funcao_probabilidade(x_vals, probabilidades)
    tabela = criar_tabela_distribuicao(x_vals, probabilidades)
    fazer_graficos(x_vals, probabilidades, titulo)
    esp, var, dp = calcular_momentos(x_vals, probabilidades)

    return esp, var, dp

# Exemplo de uso do template:
print("\nExemplo - Exercício 6: Parasitas por peixe")
# P(X=k) = 0.2 * (0.8)^k para k = 0, 1, 2, 3, 4
# P(X=5) = 1 - soma anterior

x_vals = [0, 1, 2, 3, 4, 5]
prob_base = 0.2
probabilidades = [prob_base * (0.8)**k for k in range(5)]
probabilidades.append(1 - sum(probabilidades))  # P(X=5)

esp, var, dp = resolver_exercicio_discreto(
    "Parasitas por peixe",
    x_vals,
    probabilidades
)

print("\nNOTA: Use este template para resolver os exercícios 6-20")
print("Substitua os valores de x_vals e probabilidades conforme cada problema")
```

# RESPOSTAS - DISTRIBUIÇÃO BINOMIAL

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

# ==========================================
# FUNÇÕES AUXILIARES PARA BINOMIAL
# ==========================================

def verificar_condicoes_binomial(contexto, n, p):
    """
    Verifica e explica as condições para aplicar distribuição binomial
    """
    print(f"\nCondições para Distribuição Binomial:")
    print(f"1. Número fixo de tentativas: n = {n} ✓")
    print(f"2. Cada tentativa tem apenas 2 resultados possíveis ✓")
    print(f"3. Probabilidade constante: p = {p} ✓")
    print(f"4. Tentativas independentes ✓")
    print(f"\nContexto: {contexto}")
    return True

def criar_tabela_binomial(n, p, max_k=None):
    """
    Cria tabela completa da distribuição binomial
    """
    if max_k is None:
        max_k = n

    k_vals = list(range(max_k + 1))
    probabilidades = [stats.binom.pmf(k, n, p) for k in k_vals]
    fda = [stats.binom.cdf(k, n, p) for k in k_vals]

    df = pd.DataFrame({
        'k': k_vals,
        'P(X = k)': probabilidades,
        'F(k) = P(X ≤ k)': fda
    })

    print("\nTabela da Distribuição Binomial:")
    print(df.to_string(index=False, float_format='%.6f'))
    return df

def calcular_momentos_binomial(n, p):
    """
    Calcula e valida momentos da binomial
    """
    # Momentos teóricos
    esperanca_teorica = n * p
    variancia_teorica = n * p * (1 - p)
    desvio_teorico = np.sqrt(variancia_teorica)

    print(f"\nMomentos da Binomial(n={n}, p={p}):")
    print(f"E[X] = n × p = {n} × {p} = {esperanca_teorica:.4f}")
    print(f"Var[X] = n × p × (1-p) = {n} × {p} × {1-p} = {variancia_teorica:.4f}")
    print(f"DP[X] = √Var[X] = {desvio_teorico:.4f}")

    return esperanca_teorica, variancia_teorica, desvio_teorico

def fazer_graficos_binomial(n, p, titulo):
    """
    Cria gráficos para distribuição binomial
    """
    k_vals = list(range(n + 1))
    probabilidades = [stats.binom.pmf(k, n, p) for k in k_vals]
    fda = [stats.binom.cdf(k, n, p) for k in k_vals]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # Função de probabilidade
    bars = ax1.bar(k_vals, probabilidades, alpha=0.7, color='lightgreen', edgecolor='darkgreen')
    ax1.set_xlabel('Número de sucessos (k)')
    ax1.set_ylabel('P(X = k)')
    ax1.set_title(f'Função de Probabilidade Binomial\nn={n}, p={p}\n{titulo}')
    ax1.grid(True, alpha=0.3)

    # Destacar valor com maior probabilidade
    max_idx = np.argmax(probabilidades)
    bars[max_idx].set_color('red')
    ax1.text(max_idx, probabilidades[max_idx] + 0.01, f'Moda = {max_idx}',
             ha='center', fontweight='bold')

    # FDA
    ax2.step(k_vals, fda, where='post', linewidth=2, color='blue', label='F(k)')
    ax2.scatter(k_vals, fda, color='blue', s=50, zorder=5)
    ax2.set_xlabel('k')
    ax2.set_ylabel('F(k) = P(X ≤ k)')
    ax2.set_title(f'Função de Distribuição Acumulada\nn={n}, p={p}')
    ax2.grid(True, alpha=0.3)
    ax2.set_ylim(0, 1.05)

    plt.tight_layout()
    plt.show()

def responder_perguntas_probabilidade(n, p, perguntas):
    """
    Responde perguntas específicas sobre probabilidades
    """
    print("\nRespostas às perguntas:")

    for i, pergunta in enumerate(perguntas, 1):
        print(f"\n{i}. {pergunta['texto']}")

        if pergunta['tipo'] == 'exato':
            k = pergunta['valor']
            prob = stats.binom.pmf(k, n, p)
            print(f"   P(X = {k}) = {prob:.6f}")

        elif pergunta['tipo'] == 'pelo_menos':
            k = pergunta['valor']
            prob = 1 - stats.binom.cdf(k-1, n, p)
            print(f"   P(X ≥ {k}) = 1 - P(X ≤ {k-1}) = {prob:.6f}")

        elif pergunta['tipo'] == 'no_maximo':
            k = pergunta['valor']
            prob = stats.binom.cdf(k, n, p)
            print(f"   P(X ≤ {k}) = {prob:.6f}")

        elif pergunta['tipo'] == 'menos_que':
            k = pergunta['valor']
            prob = stats.binom.cdf(k-1, n, p)
            print(f"   P(X < {k}) = P(X ≤ {k-1}) = {prob:.6f}")

        elif pergunta['tipo'] == 'entre':
            k1, k2 = pergunta['valores']
            prob = stats.binom.cdf(k2, n, p) - stats.binom.cdf(k1-1, n, p)
            print(f"   P({k1} ≤ X ≤ {k2}) = P(X ≤ {k2}) - P(X ≤ {k1-1}) = {prob:.6f}")

        elif pergunta['tipo'] == 'esperanca':
            esp = n * p
            print(f"   E[X] = n × p = {n} × {p} = {esp:.2f}")

        elif pergunta['tipo'] == 'variancia':
            var = n * p * (1 - p)
            print(f"   Var[X] = n × p × (1-p) = {var:.4f}")

# ==========================================
# EXERCÍCIO 1 - RESPOSTA BINOMIAL
# ==========================================
print("=" * 60)
print("EXERCÍCIO 1: Germinação de sementes de girassol")
print("=" * 60)

n = 12
p = 0.85

print(f"Parâmetros: n = {n} sementes, p = {p} (taxa de germinação)")

# Verificar condições
verificar_condicoes_binomial("Cada semente germina independentemente", n, p)

# Criar tabela
tabela = criar_tabela_binomial(n, p)

# Fazer gráficos
fazer_graficos_binomial(n, p, "Germinação de sementes")

# Calcular momentos
esp, var, dp = calcular_momentos_binomial(n, p)

# Responder perguntas específicas
perguntas = [
    {'tipo': 'exato', 'valor': 10, 'texto': 'Qual a probabilidade de exatamente 10 germinarem?'},
    {'tipo': 'pelo_menos', 'valor': 8, 'texto': 'Qual a probabilidade de pelo menos 8 germinarem?'},
    {'tipo': 'esperanca', 'texto': 'Quantas sementes esperamos que germinem?'}
]

responder_perguntas_probabilidade(n, p, perguntas)

print("\nInterpretação biológica:")
print(f"- Com alta taxa de germinação ({p*100}%), esperamos {esp:.1f} sementes germinando")
print(f"- É muito provável (>95%) que pelo menos 8 sementes germinem")
print(f"- A variabilidade é baixa (DP = {dp:.2f}), indicando processo confiável")

# ==========================================
# EXERCÍCIO 2 - RESPOSTA BINOMIAL
# ==========================================
print("\n" + "=" * 60)
print("EXERCÍCIO 2: Sobrevivência de filhotes de tartaruga")
print("=" * 60)

n = 15
p = 0.60

verificar_condicoes_binomial("Cada filhote tem chance independente de sobreviver", n, p)
tabela = criar_tabela_binomial(n, p)
fazer_graficos_binomial(n, p, "Sobrevivência de filhotes")
esp, var, dp = calcular_momentos_binomial(n, p)

perguntas = [
    {'tipo': 'exato', 'valor': 9, 'texto': 'P(exatamente 9 chegam ao mar) = ?'},
    {'tipo': 'menos_que', 'valor': 5, 'texto': 'P(menos de 5 chegam ao mar) = ?'},
    {'tipo': 'esperanca', 'texto': 'Qual o número mais provável de sobreviventes?'}
]

responder_perguntas_probabilidade(n, p, perguntas)

# Encontrar moda
moda = int(np.floor((n + 1) * p))
if (n + 1) * p == moda:
    print(f"\nModa: {moda} e {moda-1} (bimodal)")
else:
    print(f"\nModa: {moda} filhotes (valor mais provável)")

# ==========================================
# EXERCÍCIO 3 - RESPOSTA BINOMIAL
# ==========================================
print("\n" + "=" * 60)
print("EXERCÍCIO 3: Teste de medicamento veterinário")
print("=" * 60)

n = 20
p = 0.75

verificar_condicoes_binomial("Cada cão responde independentemente ao tratamento", n, p)
tabela = criar_tabela_binomial(n, p)
fazer_graficos_binomial(n, p, "Eficácia de medicamento")
esp, var, dp = calcular_momentos_binomial(n, p)

perguntas = [
    {'tipo': 'exato', 'valor': 20, 'texto': 'P(todos os 20 serem curados) = ?'},
    {'tipo': 'pelo_menos', 'valor': 18, 'texto': 'P(pelo menos 18 serem curados) = ?'}
]

responder_perguntas_probabilidade(n, p, perguntas)

# Intervalo de 95% (aproximação normal)
print("\nc) Intervalo de 95% dos resultados (aproximação normal):")
media = esp
desvio = dp
limite_inf = media - 1.96 * desvio
limite_sup = media + 1.96 * desvio
print(f"   Aproximadamente entre {limite_inf:.1f} e {limite_sup:.1f} cães curados")

# ==========================================
# EXERCÍCIO 4 - RESPOSTA BINOMIAL
# ==========================================
print("\n" + "=" * 60)
print("EXERCÍCIO 4: Sexo de filhotes em mamíferos")
print("=" * 60)

n = 8
p = 0.52

verificar_condicoes_binomial("Sexo de cada filhote é independente", n, p)
tabela = criar_tabela_binomial(n, p)
fazer_graficos_binomial(n, p, "Sexo de filhotes")
esp, var, dp = calcular_momentos_binomial(n, p)

# Perguntas específicas
print("\na) P(número igual de machos e fêmeas):")
prob_iguais = stats.binom.pmf(4, n, p)  # 4 machos = 4 fêmeas
print(f"   P(X = 4) = {prob_iguais:.6f}")

print("\nb) P(mais machos que fêmeas):")
prob_mais_machos = sum(stats.binom.pmf(k, n, p) for k in range(5, n+1))
print(f"   P(X > 4) = P(X ≥ 5) = {prob_mais_machos:.6f}")

print("\nc) Variância do número de machos:")
print(f"   Var[X] = {var:.4f}")

# ==========================================
# EXERCÍCIO 5 - RESPOSTA BINOMIAL
# ==========================================
print("\n" + "=" * 60)
print("EXERCÍCIO 5: Floração de orquídeas")
print("=" * 60)

n = 25
p = 0.40

verificar_condicoes_binomial("Cada planta floresce independentemente", n, p)
tabela = criar_tabela_binomial(n, p)
fazer_graficos_binomial(n, p, "Floração de orquídeas")
esp, var, dp = calcular_momentos_binomial(n, p)

perguntas = [
    {'tipo': 'exato', 'valor': 0, 'texto': 'P(nenhuma florescer) = ?'},
    {'tipo': 'entre', 'valores': (8, 12), 'texto': 'P(entre 8 e 12 florescerem) = ?'}
]

responder_perguntas_probabilidade(n, p, perguntas)

print(f"\nc) Desvio padrão esperado = {dp:.4f}")

# ==========================================
# TEMPLATE PARA EXERCÍCIOS 6-20
# ==========================================
print("\n" + "=" * 60)
print("TEMPLATE PARA EXERCÍCIOS 6-20 - BINOMIAL")
print("=" * 60)

def resolver_exercicio_binomial(titulo, n, p, contexto, perguntas_especificas):
    """
    Template para resolver qualquer exercício de distribuição binomial

    Args:
        titulo: Nome do exercício
        n: número de tentativas
        p: probabilidade de sucesso
        contexto: descrição do contexto biológico
        perguntas_especificas: lista de perguntas específicas
    """
    print(f"\n{'=' * 60}")
    print(f"{titulo}")
    print(f"{'=' * 60}")

    print(f"Parâmetros: n = {n}, p = {p}")

    # Análises padrão
    verificar_condicoes_binomial(contexto, n, p)
    tabela = criar_tabela_binomial(n, p)
    fazer_graficos_binomial(n, p, titulo)
    esp, var, dp = calcular_momentos_binomial(n, p)

    # Responder perguntas específicas
    if perguntas_especificas:
        responder_perguntas_probabilidade(n, p, perguntas_especificas)

    return esp, var, dp

# Exemplo de uso para Exercício 6:
print("\nExemplo - Exercício 6: Resistência a pesticida")

n = 18
p = 0.30
contexto = "Cada praga tem resistência independente"
perguntas = [
    {'tipo': 'exato', 'valor': 6, 'texto': 'P(exatamente 6 serem resistentes) = ?'},
    {'tipo': 'pelo_menos', 'valor': 10, 'texto': 'P(mais de 10 serem resistentes) = ?'}
]

esp, var, dp = resolver_exercicio_binomial(
    "Resistência a pesticida",
    n, p, contexto, perguntas
)

print(f"\nNúmero esperado de susceptíveis = {n} - {esp:.2f} = {n - esp:.2f}")

print("\n" + "=" * 60)
print("INSTRUÇÕES PARA DEMAIS EXERCÍCIOS:")
print("=" * 60)
print("1. Identifique n e p para cada problema")
print("2. Use o template resolver_exercicio_binomial()")
print("3. Adapte as perguntas específicas conforme o enunciado")
print("4. Sempre interprete biologicamente os resultados")
print("5. Verifique se as condições da binomial são satisfeitas")
```

# RESPOSTAS - DISTRIBUIÇÃO DE POISSON

```python


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

# ==========================================
# FUNÇÕES AUXILIARES PARA POISSON
# ==========================================

def verificar_condicoes_poisson(contexto, lambda_val, unidade_tempo_espaco):
    """
    Verifica e explica as condições para aplicar distribuição de Poisson
    """
    print(f"\nCondições para Distribuição de Poisson:")
    print(f"1. Eventos raros e independentes ✓")
    print(f"2. Taxa média constante: λ = {lambda_val} por {unidade_tempo_espaco} ✓")
    print(f"3. Probabilidade de múltiplos eventos simultâneos ≈ 0 ✓")
    print(f"4. Intervalo fixo de {unidade_tempo_espaco} ✓")
    print(f"\nContexto: {contexto}")

    # Característica única: E[X] = Var[X] = λ
    print(f"\nCaracterística da Poisson: E[X] = Var[X] = λ = {lambda_val}")
    return True

def criar_tabela_poisson(lambda_val, max_k=None):
    """
    Cria tabela da distribuição de Poisson
    """
    if max_k is None:
        # Mostrar até λ + 3√λ para cobrir ~99% da distribuição
        max_k = int(lambda_val + 3 * np.sqrt(lambda_val)) + 2

    k_vals = list(range(max_k + 1))
    probabilidades = [stats.poisson.pmf(k, lambda_val) for k in k_vals]
    fda = [stats.poisson.cdf(k, lambda_val) for k in k_vals]

    df = pd.DataFrame({
        'k': k_vals,
        'P(X = k)': probabilidades,
        'F(k) = P(X ≤ k)': fda
    })

    print(f"\nTabela da Distribuição Poisson (λ = {lambda_val}):")
    print(df.to_string(index=False, float_format='%.6f'))
    return df

def calcular_momentos_poisson(lambda_val):
    """
    Calcula momentos da distribuição de Poisson
    """
    print(f"\nMomentos da Poisson(λ = {lambda_val}):")
    print(f"E[X] = λ = {lambda_val}")
    print(f"Var[X] = λ = {lambda_val}")
    print(f"DP[X] = √λ = {np.sqrt(lambda_val):.4f}")

    # Coeficiente de variação
    cv = np.sqrt(lambda_val) / lambda_val
    print(f"CV = DP/E[X] = {cv:.4f} ({cv*100:.2f}%)")

    return lambda_val, lambda_val, np.sqrt(lambda_val)

def fazer_graficos_poisson(lambda_val, titulo):
    """
    Cria gráficos para distribuição de Poisson
    """
    max_k = int(lambda_val + 3 * np.sqrt(lambda_val)) + 2
    k_vals = list(range(max_k + 1))
    probabilidades = [stats.poisson.pmf(k, lambda_val) for k in k_vals]
    fda = [stats.poisson.cdf(k, lambda_val) for k in k_vals]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # Função de probabilidade
    bars = ax1.bar(k_vals, probabilidades, alpha=0.7, color='lightcoral', edgecolor='darkred')
    ax1.set_xlabel('Número de eventos (k)')
    ax1.set_ylabel('P(X = k)')
    ax1.set_title(f'Função de Probabilidade Poisson\nλ = {lambda_val}\n{titulo}')
    ax1.grid(True, alpha=0.3)

    # Destacar moda
    moda = int(np.floor(lambda_val))
    if moda < len(bars):
        bars[moda].set_color('red')
        ax1.text(moda, probabilidades[moda] + 0.01, f'Moda ≈ {moda}',
                 ha='center', fontweight='bold')

    # Marcar média
    ax1.axvline(lambda_val, color='blue', linestyle='--', alpha=0.7,
                label=f'E[X] = {lambda_val}')
    ax1.legend()

    # FDA
    ax2.step(k_vals, fda, where='post', linewidth=2, color='purple', label='F(k)')
    ax2.scatter(k_vals, fda, color='purple', s=50, zorder=5)
    ax2.set_xlabel('k')
    ax2.set_ylabel('F(k) = P(X ≤ k)')
    ax2.set_title(f'Função de Distribuição Acumulada\nλ = {lambda_val}')
    ax2.grid(True, alpha=0.3)
    ax2.set_ylim(0, 1.05)

    plt.tight_layout()
    plt.show()

def responder_perguntas_poisson(lambda_val, perguntas):
    """
    Responde perguntas específicas sobre distribuição de Poisson
    """
    print("\nRespostas às perguntas:")

    for i, pergunta in enumerate(perguntas, 1):
        print(f"\n{i}. {pergunta['texto']}")

        if pergunta['tipo'] == 'exato':
            k = pergunta['valor']
            prob = stats.poisson.pmf(k, lambda_val)
            print(f"   P(X = {k}) = e^(-{lambda_val}) × {lambda_val}^{k} / {k}! = {prob:.6f}")

        elif pergunta['tipo'] == 'pelo_menos':
            k = pergunta['valor']
            prob = 1 - stats.poisson.cdf(k-1, lambda_val)
            print(f"   P(X ≥ {k}) = 1 - P(X ≤ {k-1}) = {prob:.6f}")

        elif pergunta['tipo'] == 'no_maximo':
            k = pergunta['valor']
            prob = stats.poisson.cdf(k, lambda_val)
            print(f"   P(X ≤ {k}) = {prob:.6f}")

        elif pergunta['tipo'] == 'mais_que':
            k = pergunta['valor']
            prob = 1 - stats.poisson.cdf(k, lambda_val)
            print(f"   P(X > {k}) = 1 - P(X ≤ {k}) = {prob:.6f}")

        elif pergunta['tipo'] == 'menos_que':
            k = pergunta['valor']
            prob = stats.poisson.cdf(k-1, lambda_val)
            print(f"   P(X < {k}) = P(X ≤ {k-1}) = {prob:.6f}")

        elif pergunta['tipo'] == 'entre':
            k1, k2 = pergunta['valores']
            prob = stats.poisson.cdf(k2, lambda_val) - stats.poisson.cdf(k1-1, lambda_val)
            print(f"   P({k1} ≤ X ≤ {k2}) = P(X ≤ {k2}) - P(X ≤ {k1-1}) = {prob:.6f}")

        elif pergunta['tipo'] == 'zero':
            prob = stats.poisson.pmf(0, lambda_val)
            print(f"   P(X = 0) = e^(-{lambda_val}) = {prob:.6f}")

        elif pergunta['tipo'] == 'moda':
            moda = int(np.floor(lambda_val))
            print(f"   Moda ≈ {moda} (valor mais provável)")

        elif pergunta['tipo'] == 'escala':
            fator = pergunta['fator']
            novo_lambda = lambda_val * fator
            print(f"   Novo λ = {lambda_val} × {fator} = {novo_lambda}")
            print(f"   E[X] em nova escala = {novo_lambda}")

def calcular_percentis_poisson(lambda_val, percentis=[90, 95, 99]):
    """
    Calcula percentis importantes da distribuição
    """
    print(f"\nPercentis importantes (λ = {lambda_val}):")
    for p in percentis:
        valor = stats.poisson.ppf(p/100, lambda_val)
        print(f"P_{p} = {valor:.0f} (95% dos valores ≤ {valor:.0f})")

# ==========================================
# EXERCÍCIO 1 - RESPOSTA POISSON
# ==========================================
print("=" * 60)
print("EXERCÍCIO 1: Mutações em sequência de DNA")
print("=" * 60)

lambda_val = 2.5
unidade = "1000 bases"

print(f"Parâmetro: λ = {lambda_val} mutações por {unidade}")

# Verificar condições
verificar_condicoes_poisson("Mutações ocorrem aleatoriamente e independentemente",
                           lambda_val, unidade)

# Criar tabela
tabela = criar_tabela_poisson(lambda_val)

# Fazer gráficos
fazer_graficos_poisson(lambda_val, "Mutações em DNA")

# Calcular momentos
esp, var, dp = calcular_momentos_poisson(lambda_val)

# Responder perguntas específicas
perguntas = [
    {'tipo': 'zero', 'texto': 'P(nenhuma mutação) = ?'},
    {'tipo': 'exato', 'valor': 3, 'texto': 'P(exatamente 3 mutações) = ?'},
    {'tipo': 'mais_que', 'valor': 5, 'texto': 'P(mais de 5 mutações) = ?'}
]

responder_perguntas_poisson(lambda_val, perguntas)

print("\nInterpretação biológica:")
print(f"- Taxa de mutação moderada: {lambda_val} por 1000 bases")
print(f"- Probabilidade de sequência íntegra: {stats.poisson.pmf(0, lambda_val):.3f}")
print(f"- Variabilidade igual à média (característica da Poisson)")

# ==========================================
# EXERCÍCIO 2 - RESPOSTA POISSON
# ==========================================
print("\n" + "=" * 60)
print("EXERCÍCIO 2: Chegada de polinizadores")
print("=" * 60)

lambda_val = 4.2
unidade = "hora"

verificar_condicoes_poisson("Abelhas chegam independentemente", lambda_val, unidade)
tabela = criar_tabela_poisson(lambda_val)
fazer_graficos_poisson(lambda_val, "Chegada de polinizadores")
esp, var, dp = calcular_momentos_poisson(lambda_val)

perguntas = [
    {'tipo': 'zero', 'texto': 'P(nenhuma abelha chegar) = ?'},
    {'tipo': 'entre', 'valores': (3, 6), 'texto': 'P(entre 3 e 6 abelhas chegarem) = ?'},
    {'tipo': 'moda', 'texto': 'Qual o número mais provável de abelhas?'}
]

responder_perguntas_poisson(lambda_val, perguntas)

print("\nAnálise ecológica:")
print(f"- Taxa de visitação alta: {lambda_val} abelhas/hora")
print(f"- Flor é atrativa para polinizadores")
print(f"- Baixa chance de período sem visitação: {stats.poisson.pmf(0, lambda_val):.3f}")

# ==========================================
# EXERCÍCIO 3 - RESPOSTA POISSON
# ==========================================
print("\n" + "=" * 60)
print("EXERCÍCIO 3: Contagem de bactérias")
print("=" * 60)

lambda_val = 3.8
unidade = "mL"

verificar_condicoes_poisson("Bactérias distribuídas aleatoriamente", lambda_val, unidade)
tabela = criar_tabela_poisson(lambda_val)
fazer_graficos_poisson(lambda_val, "Contagem de bactérias")
esp, var, dp = calcular_momentos_poisson(lambda_val)

perguntas = [
    {'tipo': 'zero', 'texto': 'P(amostra estar estéril) = ?'},
    {'tipo': 'pelo_menos', 'valor': 5, 'texto': 'P(encontrar pelo menos 5 bactérias) = ?'},
    {'tipo': 'escala', 'fator': 5, 'texto': 'Em 5 mL, quantas bactérias esperamos?'}
]

responder_perguntas_poisson(lambda_val, perguntas)

print("\nImplicações microbiológicas:")
print(f"- Contaminação significativa: {lambda_val} bactérias/mL")
print(f"- Chance de esterilidade muito baixa: {stats.poisson.pmf(0, lambda_val):.4f}")
print(f"- Necessário tratamento antimicrobiano")

# ==========================================
# EXERCÍCIO 4 - RESPOSTA POISSON
# ==========================================
print("\n" + "=" * 60)
print("EXERCÍCIO 4: Defeitos em cromossomos")
print("=" * 60)

lambda_val = 1.8
unidade = "preparação"

verificar_condicoes_poisson("Quebras ocorrem aleatoriamente", lambda_val, unidade)
tabela = criar_tabela_poisson(lambda_val)
fazer_graficos_poisson(lambda_val, "Defeitos cromossômicos")
esp, var, dp = calcular_momentos_poisson(lambda_val)

perguntas = [
    {'tipo': 'zero', 'texto': 'P(cromossomo íntegro) = ?'},
    {'tipo': 'exato', 'valor': 2, 'texto': 'P(exatamente 2 quebras) = ?'},
    {'tipo': 'mais_que', 'valor': 4, 'texto': 'P(mais de 4 quebras) = ?'}
]

responder_perguntas_poisson(lambda_val, perguntas)

print("\nAnálise citogenética:")
print(f"- Taxa moderada de quebras: {lambda_val} por preparação")
print(f"- Chance de integridade: {stats.poisson.pmf(0, lambda_val):.3f}")
print(f"- Pode indicar estresse celular ou agente mutagênico")

# ==========================================
# EXERCÍCIO 5 - RESPOSTA POISSON
# ==========================================
print("\n" + "=" * 60)
print("EXERCÍCIO 5: Nascimento de gêmeos")
print("=" * 60)

lambda_val = 0.8
unidade = "mês"

verificar_condicoes_poisson("Partos gemelares são eventos raros", lambda_val, unidade)
tabela = criar_tabela_poisson(lambda_val)
fazer_graficos_poisson(lambda_val, "Nascimento de gêmeos")
esp, var, dp = calcular_momentos_poisson(lambda_val)

perguntas = [
    {'tipo': 'zero', 'texto': 'P(nenhum parto gemelar) = ?'},
    {'tipo': 'exato', 'valor': 1, 'texto': 'P(exatamente 1 parto gemelar) = ?'},
    {'tipo': 'pelo_menos', 'valor': 3, 'texto': 'P(3 ou mais partos gemelares) = ?'}
]

responder_perguntas_poisson(lambda_val, perguntas)

print("\nAnálise demográfica:")
print(f"- Taxa baixa de gemelaridade: {lambda_val} por mês")
print(f"- Evento relativamente raro na população")
print(f"- Meses sem gêmeos são comuns: {stats.poisson.pmf(0, lambda_val):.3f}")

# ==========================================
# TEMPLATE PARA EXERCÍCIOS 6-20
# ==========================================
print("\n" + "=" * 60)
print("TEMPLATE PARA EXERCÍCIOS 6-20 - POISSON")
print("=" * 60)

def resolver_exercicio_poisson(titulo, lambda_val, unidade, contexto, perguntas_especificas):
    """
    Template para resolver qualquer exercício de distribuição de Poisson

    Args:
        titulo: Nome do exercício
        lambda_val: parâmetro lambda (taxa média)
        unidade: unidade de tempo/espaço
        contexto: descrição do contexto biológico
        perguntas_especificas: lista de perguntas específicas
    """
    print(f"\n{'=' * 60}")
    print(f"{titulo}")
    print(f"{'=' * 60}")

    print(f"Parâmetro: λ = {lambda_val} por {unidade}")

    # Análises padrão
    verificar_condicoes_poisson(contexto, lambda_val, unidade)
    tabela = criar_tabela_poisson(lambda_val)
    fazer_graficos_poisson(lambda_val, titulo)
    esp, var, dp = calcular_momentos_poisson(lambda_val)

    # Responder perguntas específicas
    if perguntas_especificas:
        responder_perguntas_poisson(lambda_val, perguntas_especificas)

    # Calcular percentis
    calcular_percentis_poisson(lambda_val)

    return esp, var, dp

# Exemplo de uso para Exercício 6:
print("\nExemplo - Exercício 6: Parasitas em peixes")

lambda_val = 2.3
unidade = "peixe"
contexto = "Parasitas se aderem independentemente"
perguntas = [
    {'tipo': 'zero', 'texto': 'P(peixe sem parasitas) = ?'},
    {'tipo': 'entre', 'valores': (1, 3), 'texto': 'P(peixe com 1 a 3 parasitas) = ?'}
]

esp, var, dp = resolver_exercicio_poisson(
    "Parasitas em peixes",
    lambda_val, unidade, contexto, perguntas
)

print("\nPercentil 95%:")
p95 = stats.poisson.ppf(0.95, lambda_val)
print(f"95% dos peixes têm ≤ {p95:.0f} parasitas")
print(f"Apenas 5% têm > {p95:.0f} parasitas (casos severos)")

print("\n" + "=" * 60)
print("INSTRUÇÕES PARA DEMAIS EXERCÍCIOS:")
print("=" * 60)
print("1. Identifique λ (taxa média) para cada problema")
print("2. Determine a unidade de tempo/espaço")
print("3. Use o template resolver_exercicio_poisson()")
print("4. Verifique se as condições da Poisson são satisfeitas:")
print("   - Eventos raros e independentes")
print("   - Taxa constante")
print("   - Intervalo fixo")
print("5. Lembre-se: E[X] = Var[X] = λ (característica única)")
print("6. Interprete biologicamente os resultados")

print("\n" + "=" * 60)
print("DICAS IMPORTANTES:")
print("=" * 60)
print("• Para λ pequeno (< 1): maioria dos valores será 0 ou 1")
print("• Para λ médio (1-5): distribuição moderadamente dispersa")
print("• Para λ grande (> 10): aproxima-se da normal")
print("• Moda ≈ floor(λ) para λ não inteiro")
print("• Poisson é limite da Binomial quando n→∞, p→0, np→λ")
```
