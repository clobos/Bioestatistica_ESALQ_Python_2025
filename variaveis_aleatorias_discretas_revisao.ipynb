# Número de ovos em ninhos de aves

Uma espécie de ave põe entre 2 a 8 ovos por ninho


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Definir valores e probabilidades
valores = [2,3,4,5,6,7,8]
probabilidades = [0.05,0.15,0.25,0.3,0.15,0.08,0.02]

# Criar distribuição personalizada
class DistribuicaoPersonalizada:
    def __init__(self, valores, probabilidades):
        self.valores = np.array(valores)
        self.probs = np.array(probabilidades)

    def pmf(self, x):
        """Função de probabilidade"""
        if x in self.valores:
            index = np.where(self.valores == x)[0][0]
            return self.probs[index]
        return 0.0

    def cdf(self, x):
        """Função de distribuição acumulada"""
        return np.sum(self.probs[self.valores <= x])

# Criar instância da distribuição
dist = DistribuicaoPersonalizada(valores, probabilidades)

# Exemplos de cálculo
x = 4  # Exemplo: valor 4
print(f"P(X = {x}) = {dist.pmf(x):.4f}")
print(f"P(X ≤ {x}) = {dist.cdf(x):.4f}")

# Plotar PMF
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(valores, probabilidades, alpha=0.8, color='#1FB8CD')
plt.xlabel('Valor (x)')
plt.ylabel('P(X = x)')
plt.title('Função de Probabilidade (PMF)')
plt.grid(True, alpha=0.3)

# Plotar CDF
plt.subplot(1, 2, 2)
cdf_values = [dist.cdf(v) for v in valores]
plt.step(valores, cdf_values, where='post', linewidth=2, color='#FFC185')
plt.xlabel('Valor (x)')
plt.ylabel('F(x) = P(X ≤ x)')
plt.title('Função de Distribuição Acumulada (CDF)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
